```python
#Import libraries
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By

import pandas as pd
```

```python
#Fetch the webpage
url = "https://mysg-s5.mpl.mobilelegends.com/stats/"  # Replace with the actual URL
response = requests.get(url)
```

```python
headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.0.0 Safari/537.36'
}
```

```python
#Parse the HTML
soup = BeautifulSoup(response.content, 'html')
```

### Regular Season


```python
#Find the table
table_player = soup.find('table', id = 'tablepress-15')
print(table_player)
```

```python
# looking for title and get the text onlt
title_player = table_player.find_all('th')
player_table_title = [title.text for title in title_player]

print(player_table_title)
```

```python
# change to dataframe
df = pd.DataFrame(columns = player_table_title)
df
```

```python
# Extract the data n put in data frame
column_player = table_player.find_all('tr')

for row in column_player[1:]:
    cols = row.find_all('td')
    data = [col.text.strip() for col in cols]

    length = len(df)
    df.loc[length] = data


df
```

```python
df.to_csv('regular_player.csv', index=False)
```
